## Videos

```python
n_frames = 500
interval = 40
steps_per_frame = 2
rotations = 2

colors_ = [ps.CMAP(i / 19.) for i in range(20)]
```

```python
def make_colors(color, steps, a=5., b=0.01, alphamin=0.2):
    rgba = np.zeros((steps, 4))
    rgba[:, :] = colors.to_rgba(color)
    x = np.linspace(-steps + 1, 0, steps)
    alphas = alphamin + (1. - alphamin)/(1 + np.exp(-a - b*x)) 
    rgba[:, 3] = alphas
    return rgba[1:]

def make_segments(i, steps):
    data = trajectories2[i][:steps, :].reshape((-1, 1, 3))  # (steps, 1, 3)
    segments = np.hstack([data[:-1], data[1:]])  # (steps, 2, 3)
    return segments

def make_line_collection(i, steps, color, lw=1.0, ls="-"):
    segments = make_segments(i, steps)
    colors = make_colors(color, steps)
    lc = Line3DCollection(segments=segments, colors=colors, linewidths=lw, linestyles=ls, zorder=-1.)
    return lc
```

```python
fig, ax = ps.figure(height=0.5*ps.TEXTWIDTH, cbar=False, make3d=True, large_margin=0.05, small_margin=0.05)

ax.set_xlim3d(-3,3)
ax.set_ylim3d(-3,3)
ax.set_zlim3d(-3,3)
ax.set_xticks([-2., 0., 2.])
ax.set_yticks([-2., 0., 2.])
ax.set_zticks([-2., 0., 2.])
ax.set_xticklabels(["", "", ""])
ax.set_yticklabels(["", "", ""])
ax.set_zticklabels(["", "", ""])
ax.set_xlabel("$x_0$", labelpad=-5.)
ax.set_ylabel("$x_1$", labelpad=-5.)
ax.set_zlabel("$x_2$", labelpad=-5.)

ax.view_init(elev=20, azim=-70)

lcs, points = [], []
for i, color in enumerate(colors_):
    lcs.append(make_line_collection(i, 1, color))
    ax.add_collection(lcs[-1])
    
    points.append(
        ax.plot(
            trajectories2[i][:1, 0], trajectories2[i][:1, 1], trajectories2[i][:1, 2],
            linestyle= " ", marker="o", markersize=7.,c=color
        )[0]
    )

def update(t):
    if t % 50 == 0:
        print(t)
    length = t * steps_per_frame
    
    for i, (lc, color) in enumerate(zip(lcs, colors_)):
        lc.set_segments(make_segments(i, length))
        lc.set_color(make_colors(color, length))
        
    for trajectory, point in zip(trajectories2, points):
        point.set_xdata(trajectory[length - 1:length, 0])
        point.set_ydata(trajectory[length - 1:length, 1])
        point.set_3d_properties(trajectory[length - 1:length, 2])
    
    ax.view_init(elev=20, azim=30 + t * rotations * 360. / n_frames)

anim = FuncAnimation(fig, update, frames=np.arange(0, n_frames), interval=interval)
anim.save('../figures/lorenz_system_random.gif', dpi=100, writer='imagemagick')
```
